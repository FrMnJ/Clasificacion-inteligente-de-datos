# Perceptrón
## Fundamentos de la técnica
El perceptrón es la unidad fundamental de una red neuronal. Está compuesto por cuatro elementos principales: entradas, pesos y un bias, una suma ponderada y una función de activación. Cada entrada se multiplica por su peso correspondiente, se suma el bias y, finalmente, la función de activación determina la salida del perceptrón (Bhardwaj, 2020).

## Modelo Matemático del Perceptrón

El perceptron consiste en:
 - Un conjunto de entradas $x_1, x_2, x_3, \dots, x_n$.
 - Un conjunto de pesos $w_1, w_2, w_3, \dots, w_n$
 - Un función de suma ponderada $\sum w_ix_i+b$ donde $b$ es el sesgo o bias.
 - Una función de activación que determina la salida.


## Pipeline

### Feature Engineering
Las variables de entrada describen las señales de la compuerta logica AND $X$ y su salida respectivamente $Y$.





```python
# Librerías utilizadas:
# - pandas (pd): manejo y exploración de datos en estructuras DataFrame.
# - numpy (np): arrays n-dimensionales y operaciones numéricas eficientes.
# - matplotlib.pyplot (plt): visualización de datos (gráficas 2D).
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Conjunto de datos de la compuerta lógica AND
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([0,0,0,1])

# Construcción de un DataFrame para inspección rápida
df = pd.DataFrame(X, columns=['x1', 'x2'])
df['y'] = Y

print(df.describe())
print()

# Scatter plot para visualizar separabilidad lineal
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

### Model Selection
Para emplear un clasificador lineal, y específicamente un perceptrón simple, es necesario que los datos sean linealmente separables (Chandra, 2018). Otras razones para elegir un perceptrón incluyen su alta interpretabilidad y su bajo costo computacional, lo que lo hace eficiente para tareas de clasificación básicas (Serengil, 2020).

### Model Training

```python
# Entrenamiento del perceptrón:
# - sklearn.linear_model.Perceptron: clase que implementa el algoritmo perceptrón para clasificación lineal.
# - Parámetros importantes: random_state (aleatoriedad reproducible), max_iter (máximo de iteraciones), tol (tolerancia de parada).
from sklearn.linear_model import Perceptron
# Crear el modelo con semilla para reproducibilidad
model = Perceptron(random_state=1)
# Ajustar pesos y bias usando los datos X (entradas) y Y (etiquetas)
model.fit(X, Y)
```

### Prediction

```python
# Función de prueba del perceptrón:
# - test_perceptron(model, X, Y): recorre cada entrada X[i], obtiene la predicción y la compara con la etiqueta esperada Y[i].
# - model.predict espera un array 2D de muestras; aquí pasamos una muestra a la vez [X[i]].
def test_perceptron(model, X, Y):
    for i in range(len(X)):
        prediction = model.predict([X[i]])
        expected = Y[i]
        print(f"Entrada: {X[i]}, Predicción: {prediction[0]}, Esperado: {expected}")

# Ejecutar las pruebas de predicción para todas las muestras
test_perceptron(model, X, Y)
```

### Model Evaluation

```python
# Evaluación del modelo:
# - sklearn.metrics.accuracy_score: devuelve la fracción de predicciones correctas (exactitud).
from sklearn.metrics import accuracy_score
y_pred = model.predict(X)
accuracy_score(Y, y_pred)
```
